```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

## Single Chart

```python
x = np.linspace(-10, 10, 100)
y = x**2

# Create figure and Axes
fig, ax = plt.subplots()

# Draw the chart includes 2 axis (Ox, Oy) on Axes
ax.plot(x, y, label='y = x^2', color='red')

# title, label in Axes
ax.set_title("Đồ thị hàm số y = x^2")
ax.set_xlabel("x")
ax.set_ylabel("y")

# Thêm lưới và note
ax.grid(True)
ax.legend()

# Show
plt.show()
```

## Multiple Chart

```python
x = np.linspace(-10, 10, 100)
y = x**2

# Tao 1 figure voi 2 subplot (1 hang, 2 cot)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5)) # figsize giup tang kich thuoc tong the

# Subplot 1
ax1.plot(x, y, label='y = x^2', color="green")
ax1.set_title("Subplot 1: Do thi y = x^2")
ax1.set_xlabel("x")
ax1.set_ylabel("y")
ax1.grid(True)
ax1.legend()

# Subplot 2
ax2.plot(x, -y, label='y = -x^2', color="red")
ax2.set_title("Subplot 2: Do thi y = -x^2")
ax2.set_xlabel("x")
ax2.set_ylabel("y")
ax2.grid(True)
ax2.legend()

plt.show()
```

## Line Chart

```python
x = np.linspace(-10, 10, 100)
y = x**2

plt.plot(x, y, label = 'y = x^2', color='r', linewidth=2, linestyle='--')

plt.title('Line Chart: y = x^2')
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()
plt.show()
```

## Boxplot Chart

```python
# sample data
np.random.seed(0)
data = [np.random.normal(loc=mean, scale=5, size=100) for mean in [10, 20, 30]]

# boxplot
plt.boxplot(data, labels=["Nhom 1", "Nhom 2", "Nhom 3"])
plt.title("Boxplot: So sanh 3 nhom du lieu")
plt.ylabel("Gia tri")

plt.show()
```

## Histogram Chart

```python
np.random.seed(42)
data = np.random.normal(loc=0, scale=1, size=1000)

plt.hist(data, bins=30, color='skyblue', edgecolor='r')

plt.title("Histogram: Phân phối dữ liệu chuẩn")
plt.xlabel("Giá trị")
plt.ylabel("Tần suất")

plt.show()
```

# Seaborn

```python
n = 200
df = pd.DataFrame({
    'age': np.random.randint(18, 60, size=n),
    'income': np.random.normal(loc=50000, scale=12000, size=n),
    'gender': np.random.choice(['Male', 'Female'], size=n),
    'category': np.random.choice(['A', 'B', 'C'], size=n),
    'score': np.random.uniform(0, 100, size=n),
    'purchased': np.random.choice([0, 1], size=n, p=[0.4, 0.6])
})
```

```python
df.head()
```

## Histogram & KDE (Kernel Density Estimate)

```python
plt.figure(figsize=(10, 5))
sns.histplot(df['score'], bins=20, kde=True, color='skyblue')

plt.title('Histogram & KDE of score')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(8,5))
sns.kdeplot(df['score'], color='b', fill=True)

plt.title('KDE của cột score')
plt.xlabel('Score')
plt.ylabel('Mật độ')

plt.grid(True)
plt.show()
```

## Scatter

- Scatter plot (biểu đồ phân tán) dùng để thể hiện mối quan hệ giữa hai biến số liên tục. 

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df['age'], y=df['income'], hue=df['gender'], palette='Set1')

plt.title('Scatter Plot: Age vs Income')
plt.xlabel('Age')
plt.ylabel('Income')

plt.show()
```

## Violin
- Violin plot là sự kết hợp giữa: boxplot(biểu đồ hộp - hiển thị phân vị, trung vị, và độ trải dài của dữ liệu), kde plot(đường mật độ - thể hiện hình dạng phân phối của dữ liệu).
- Violin plot dùng để hiển thị phân phối của một biến số liên tục (như thu nhập, điểm số, ...) theo từng nhóm (như giới tính, vùng miền, ...)
- Violin plot cho bạn biết: 
    + Trung vị, phần lớn dữ liệu nằm ở đâu.
    + Dữ liệu có phân bố lệch hay không.
    + Dữ liệu có nhiều đỉnh (multi-modal) không.

```python
plt.figure(figsize=(10, 5))
sns.violinplot(x=df['category'],
               y=df['income'],
               palette='muted')

plt.title('Violin plot: Income phân theo Category')
plt.xlabel('Category')
plt.ylabel('Income')

plt.show()
```

## Heatmap

```python
df = df[['age', 'income', 'score', 'purchased']]

correlation_matrix = df.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, 
            annot=True,
            cmap='coolwarm',
            fmt='.2f', linewidth=0.5)

plt.title('Heatmap: Ma trận tương quan')
plt.show()
```
