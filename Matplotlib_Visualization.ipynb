```python
import matplotlib.pyplot as plt
import numpy as np
```

# Bai 1

```python
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
temperatures = [22, 24, 25, 23, 26, 28, 27]

plt.figure(figsize=(8, 5))
plt.plot(days, temperatures, color='red', linestyle='--', marker='o', linewidth=2, markersize=8)
plt.title("Weekly Temperature")
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.grid(True)
plt.show()
```

# Bai 2

```python
# create dataset
data = np.random.normal(0, 1, 1000)

# draw histogram
plt.figure(figsize=(8, 5))
plt.hist(data, bins=30, color='green',edgecolor='black', density=True, alpha=0.7)
plt.title('Normal Distribution Histogram')
plt.xlabel('Value')
plt.ylabel('Density')

plt.show()
```

# Bai 3

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Subplot 1
ax1.plot(x, y1, 'b-', linewidth=2)
ax1.set_title('Sine Function', fontsize=14)
ax1.set_xlabel('x', fontsize=12)
ax1.set_ylabel('sin(x)', fontsize=12)

# Subplot 2
ax2.plot(x, y2, 'r-', linewidth=2)
ax2.set_title('Cosine Function')
ax2.set_xlabel('x')
ax2.set_ylabel('cos(x)')

plt.tight_layout(pad=3)
plt.show()
```

# Bai 4

```python
quarters = ['Q1', 'Q2', 'Q3', 'Q4']
sales_2022 = [120, 135, 145, 160]
sales_2023 = [130, 148, 155, 170]

x = np.arange(len(quarters))
width = 0.35

fig, ax = plt.subplots(figsize=(9, 6))
rects1 = ax.bar(x - width/2, sales_2022, width,
                label='2022', color='skyblue', edgecolor='black')
rects2 = ax.bar(x + width/2, sales_2023, width,
                label='2023', color='salmon', edgecolor='black')

ax.set_title('Quarterly Sales Comparison', fontsize=16)
ax.set_xlabel('Quarter', fontsize=12)
ax.set_ylabel('Sales (million $)', fontsize=12)
ax.set_xticks(x)
ax.set_xticklabels(quarters)
ax.legend(fontsize=12)

def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate(f'{height}',
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)

plt.tight_layout()
plt.show()
```
