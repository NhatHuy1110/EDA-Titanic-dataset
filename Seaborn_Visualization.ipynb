# Seaborn 
- Là thư viện visualization nâng cao xây dựng trên Matplotlib.
- Giao diện đơn giản cho visualization statistic.
- Tích hợp tốt với Pandas DataFrames.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset("tips")
tips
```

## Bai 1

```python
plt.figure(figsize=(10, 5))
sns.histplot(data=tips, x='total_bill', kde=True, bins=20, color='skyblue', edgecolor='navy')
plt.title('Total bill Distribution')
plt.xlabel('Total bill')
plt.ylabel('Frequency')
```

```python
plt.figure(figsize=(10, 5))
sns.violinplot(data=tips, x='time', y='total_bill',
                palette={'Lunch':'lightgreen', 'Dinner':'salmon'})
plt.title('Phan phoi hoa don theo bua an')
plt.xlabel('Thoi diem')
plt.ylabel('Total Bill')

plt.tight_layout()
plt.show()
```

## Bai 2

```python
diamonds = sns.load_dataset("diamonds").sample(1000) # Giam kich thuoc dataset
```

```python
diamonds
```

```python
plt.figure(figsize=(12, 6))

sns.lmplot(
    data = diamonds,
    x='carat',
    y='price',
    hue='cut',
    col='cut',
    col_wrap=3,
    scatter_kws={
        's': diamonds['depth'].mean()*2,
        'alpha':0.6
    },
    line_kws={
        'lw':2
    },
    height=4,
    aspect=1.2
)

plt.suptitle('Moi quan he Carat - Price theo chat luong kim cuong', y=1.02)
plt.tight_layout()
plt.show()
```

## Bai 3

```python
flights = sns.load_dataset("flights")
```

```python
# Heatmap
flights_pivot = flights.pivot(index="year", columns="month", values="passengers")
plt.figure(figsize=(12, 6))
sns.heatmap(flights_pivot, annot=True, fmt="d",
           cmap="YlOrRd", linewidths=.5)
plt.title('So luong hang khach theo thang/nam')
```

```python
# FaceGrid
g = sns.FacetGrid(flights, col="month", col_wrap=4, height=3)
g.map(sns.lineplot, "year", "passengers", color="steelblue")
g.set_axis_labels("Năm", "Số khách")
g.fig.suptitle('Trend hành khách theo tháng (1949-1960)', y=1.02)

plt.tight_layout()
plt.show()
```
